```python
# Day 3, Problem 1
file = open('day3input.txt', 'r')
content = file.read()
diagnostic = content.strip().split('\n')

# lists to combine later
gamma_ = []
epsilon_ = []

# each bit
bit1 = 0
bit2 = 0
bit3 = 0
bit4 = 0
bit5 = 0
bit6 = 0
bit7 = 0
bit8 = 0
bit9 = 0
bit10 = 0
bit11 = 0
bit12 = 0

# 12 bits each, 1000 bytes
for code in diagnostic:
    bit1 += int(code[0])
    bit2 += int(code[1])
    bit3 += int(code[2])
    bit4 += int(code[3])
    bit5 += int(code[4])
    bit6 += int(code[5])
    bit7 += int(code[6])
    bit8 += int(code[7])
    bit9 += int(code[8])
    bit10 += int(code[9])
    bit11 += int(code[10])
    bit12 += int(code[11])
    
diag_bits = [bit1,bit2,bit3,bit4,bit5,bit6,bit7,bit8,bit9,bit10,bit11,bit12]

for val in diag_bits:
    if val > 500:
        gamma_.append('1')
        epsilon_.append('0')
    else:
        gamma_.append('0')
        epsilon_.append('1')
        
# int(x,2) converts binary to decimal by specifying base 2
gamma = int("".join(gamma_),2)
epsilon = int("".join(epsilon_),2)

product = gamma * epsilon
product
# 3885894 - correct!
# probably a way to represent those bits with less hardcoding
```
